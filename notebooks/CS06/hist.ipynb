```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```
tips_data = sns.load_dataset('tips')
tips_data.head()
```

```
sns.distplot(tips_data.total_bill)
plt.show()
```

```
sns.distplot(tips_data.total_bill, rug=True)
plt.show()
```

```
sns.distplot(tips_data.total_bill, kde=None)
plt.show()
```

```
sns.distplot(tips_data[tips_data.smoker == 'Yes'], rug=True)
sns.distplot(tips_data[tips_data.smoker == 'No'], rug=True)
plt.legend()
plt.show()
```

```
sns.pairplot(tips_data)
```
