```
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```
edu_data = pd.read_csv('http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv')
```

```
edu_data.head()
```

```
edu_data.plot(x='Year', y='Agriculture', figsize=(15, 10))
```

```
edu_data.loc[0, edu_data.iloc[0] < 20]
```

```
under_20_pct = edu_data.loc[0, edu_data.iloc[0] < 20].index
```

```
edu_data.plot(x='Year', y=under_20_pct, figsize=(15, 10))
```
