```
from vega_datasets import data
import matplotlib.pyplot as plt
import seaborn as sns
```

```
cars = data.cars()
cars.head()
```

```
plt.scatter(cars.Horsepower, cars.Miles_per_Gallon)
plt.xlabel("Horse Power")
plt.ylabel("Miles per Gallon")
plt.title("Miles per Gallon vs Horsepower")
plt.show()
```

```
for region in cars.Origin.unique():
  data_sub = cars[cars.Origin == region]
  plt.scatter(data_sub.Horsepower, data_sub.Miles_per_Gallon, label=region)

plt.xlabel("Horse Power")
plt.ylabel("Miles per Gallon")
plt.title("Miles per Gallon vs Horsepower per Origin")
plt.legend()
plt.show()
```

```
sns.scatterplot(x="Horsepower", y="Miles_per_Gallon", hue='Origin', size='Cylinders', alpha=0.5, data=cars)

for i, car in enumerate(cars.Name):
  if cars.Horsepower[i] > 125 and cars.Miles_per_Gallon[i] > 30:
    plt.annotate(car, (cars.Horsepower[i], cars.Miles_per_Gallon[i]))

plt.xlabel("Horse Power")
plt.ylabel("Miles per Gallon")
plt.title("Miles per Gallon vs Horsepower per Origin per Cylinders")
plt.legend()
plt.show()
```

```
sns.lmplot(x="Horsepower", y="Miles_per_Gallon", data=cars, fit_reg=True)

plt.title("Regression Line: Miles per Gallon vs Horsepower")
plt.xlabel("Miles per Gallon")
plt.ylabel("Horse Power")
plt.show()
```
